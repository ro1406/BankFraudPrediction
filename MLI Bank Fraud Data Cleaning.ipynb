# MLI Bank Fraud Prediction Data Cleaning

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
df=pd.read_csv("/machine_learning_india_ai_challenge-dataset/TRAIN - CLEANED.csv")
```

```python
original=df
```

```python
df=df.drop(columns=['Index','Class'])
```

```python
df
```

```python
df=df.apply(stats.zscore)
```

```python
df
```

```python
original
```

```python
cols=df.columns
for i in cols:
    ind=df[ ((df[i] >=3) | (df[i]<=-3)) & (original['Class']!=1) ].index
    df.drop(ind,  inplace=True)
    original.drop(ind,inplace=True)
    #df.reset_index()
    
print(len(df))
print(len(original))
```

```python
original=original.drop(columns=['Index'])
```

```python
original.to_csv("/machine_learning_india_ai_challenge-dataset/CLEANED Training data - Python.csv",index=False)
```
